```python
import json
import numpy as np
```

```python
file = open('C:\\Users\\Vignesh\\Desktop\\E-Book Project\\resources\\embeddings\\book_embeddings.ndjson', 'r')
book_embeddings = json.loads(file.read())
file.close()

file = open('C:\\Users\\Vignesh\\Desktop\\E-Book Project\\resources\\json\\book_index', 'r')
book_index  = json.loads(file.read())
file.close()

file = open('C:\\Users\\Vignesh\\Desktop\\E-Book Project\\resources\\json\\index_book', 'r')
index_book = json.loads(file.read())
file.close()
```

```python
li = list(book_index.keys())
li.sort()
li[200:]
```

```python
book_embeddings = np.array(book_embeddings, dtype = float)
```

```python
book_embeddings
```

```python
def recommendation(book, book_embeddings):
    
    dist = np.dot(book_embeddings, book_embeddings[book_index[book]])
    sort_dist = np.argsort(dist)
    
    recommendation = []
    for i in np.flipud(sort_dist[-10:]):
        recommendation.append([index_book[str(i)], dist[i]])
    return recommendation       
        
```

```python
recommendation('Blood Beast/Darren Shan', book_embeddings)
```
